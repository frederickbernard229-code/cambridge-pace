**Activity 1.2.5 Response**

- **Repeat purchase rate** — This is one of the clearest signs of loyalty and long-term customer value. It shows whether customers come back voluntarily, which is often more informative than a one-time purchase.

- **Purchase amount** — This helps identify how much customers spend and whether different segments contribute more value. It is useful for spotting high-value customers, pricing sensitivity, and changes in spending over time.

- **Conversion rate** — This shows how effectively browsing interest turns into actual purchases. It is a strong indicator of friction in the buying journey and helps separate interested visitors from committed customers.

- **Cart abandonment rate** — This is especially useful for understanding where customers drop out before buying. A high abandonment rate often signals checkout friction, pricing surprises, or trust issues.

- **Customer satisfaction score** — Satisfaction is a strong signal of how customers feel about the experience, and it often predicts retention and advocacy. It helps explain not just what customers do, but whether they are likely to stay engaged.

- **Customer support response time** — Service speed can strongly affect churn, repeat buying, and overall satisfaction. If customers wait too long for help, that often becomes a direct driver of negative behaviour and lost business.
