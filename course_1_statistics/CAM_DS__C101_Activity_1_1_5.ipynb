# Activity 1.1.5: Communik8 Customer Feedback Analysis

**Contact Centre.** Customers experience long wait times and receive responses that fail to answer their queries. *Impact:* retention (customers churn at the moment of friction), operational cost (repeat calls), and brand damage. *Consequence of inaction:* the contact centre is the safety net when other channels fail — when it's broken, every other failure gets amplified.

**Product Delivery.** Products are delivered later than promised. *Impact:* new-customer acquisition (orders cancelled pre-relationship), wasted Customer Acquisition Cost (CAC), refund processing cost, and early-lifecycle brand damage. *Consequence of inaction:* directly tied to the "reduction in new customers" line in the scenario — customers leaving before the relationship has begun.

**Store Staff Service.** Store staff do not solve customer queries effectively and are rude. *Impact:* in-store conversion, retention, and local reputation. *Consequence of inaction:* stores are fixed-cost investments — staff who drive customers away destroy property-level economics, and physical-store customers tend to have the highest lifetime value.

**Website.** Website information does not answer customer queries. *Impact:* digital conversion, customer acquisition cost (broken self-service pushes demand into expensive channels), and digital/SEO brand. *Consequence of inaction:* the website is usually the cheapest channel — when it fails, everything else gets more expensive.

**Chatbot.** Responses are unhelpful and fail to answer the query, and are often grammatically incorrect, reducing clarity. *Impact:* same as website (deflection, cost, customer experience), plus social media risk — bad chatbot responses get screenshotted. *Consequence of inaction:* chatbots speak in the brand's voice; a bad one is the brand looking foolish in public.

**Network Coverage.** Customers are unable to make calls or use mobile internet. *Impact:* retention (highest churn risk on the list — customers leave at contract end), Ofcom service-quality exposure, and mis-selling claims. *Consequence of inaction:* a telco that can't connect calls has no product.

**Cancellation / Contract Terms.** Cooling-off cancellation requests are not actioned, locking customers into 24-month contracts against their will. *Impact:* Ofcom enforcement and fines, Ombudsman referrals, mis-selling/class-action risk, and reputational damage. *Consequence of inaction:* this is a legal failure, not just a customer-experience one — cooling-off is a statutory right under the Consumer Rights Act and Ofcom rules.

**Complaint Handling.** Complaints sent by email receive no response, leaving customers in limbo and prompting cancellations. *Impact:* Ofcom complaints-code compliance, Ombudsman per-case fees, and retention (the 10-year loyal customer is leaving over this). *Consequence of inaction:* in regulated industries, complaint handling is itself regulated — and resolved complainers retain better than customers who never complained, making this a loyalty opportunity rather than just a cost.

**Billing and Collections.** Customers are erroneously billed and then harassed by collections agents. *Impact:* Ofcom billing-accuracy rules, debt-collection conduct rules, data protection / credit-file claims, customer compensation, and viral reputational damage. *Consequence of inaction:* pursuing a customer for your own billing error, via a debt collector, is the kind of story that makes national press.

**Brand Sentiment.** Customers are losing confidence in the brand and openly recommending alternatives. *Impact:* customer acquisition cost (marketing has to overcome negative sentiment), premium-pricing erosion, and word-of-mouth in a switch-easy market. *Consequence of inaction:* brand sentiment is the multiplier on everything else. Fixing it requires upstream fixes — but proactive recovery (responsive complaint handling, advocacy) can lift Net Promoter Score (NPS) faster than fixing things silently.
