# Getting started

The document that you're reading is not a static web page, but an interactive environment called a <strong>Colab notebook</strong> that lets you write and execute code.

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

When using Colab Notebook, we encourage you to adhere to certain best practices to ensure optimal efficiency and effectiveness.

Here are some tips and best practices to help you stay organised:

**Organise your code clearly:**

*   Use code cells to logically group your operations. For example, separate data loading, data cleaning, data visualisation, and model training.
*   Use text cells (Markdown) to explain your decisions, rationale, and findings and to write down important information.
*   Use descriptive names in lower case, and separate words with underscores (e.g. data_clean) when naming variables and functions. For constants, use upper case with underscores (e.g. TAX_RATE). Avoid single-letter names unless in a local context (e.g. loop indices).
*   Use four spaces per indentation level, and limit all lines to 80 characters.
*   Group your imports at the beginning of the file. Follow the order: standard libraries, third-party libraries, and local modules.
*   To clarify the code, you can insert comments. However, inline comments should be used sparingly and only used where relevant. Use docstrings to describe the purpose, parameters, and return value.

**Data science best practices**:
*   **Reproducibility:** Ensure your code can be run from top to bottom without errors in a fresh environment. Use random seeds where randomness is involved.
*   **Data management:** Use Pandas or other suitable libraries for data manipulation. Always inspect your data before performing operations.
*  **Visualisations:** Use libraries like Matplotlib and Seaborn for visualisations. Always provide labels for axes, legends, and a title for all plots.
*   **Dependencies:** If using any non-standard libraries, document them. In Google Colab, you can use !pip freeze to list all installed packages.
*  **Python style guide:** Ensure you are familiar with the PEP 8 – [Style Guide for Python Code.](https://peps.python.org/pep-0008/)
